## ***Step 6***: Mapping the target variable in numeric values

```python
# Mapping the variables
data['Species']=data['Species'].map({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
data.head()
```

```python
plt.style.use('ggplot')
sns.heatmap(data.corr(),annot=True)
```

## ***Step 7***: importing required Machine Learning libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn import tree
```

## ***Step 8***: Data Spliting of the target variables

```python
#split dataset in features and target variable
X = data.drop('Species',axis='columns')
y = data['Species']
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=12)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## ***Step 9***: Decision Tree Model Training

```python
# instantiate the classifier 
classifier = DecisionTreeClassifier()

# fit the model
classifier.fit(X_train, y_train)
```

```python
#Predicting the test set result  
y_pred= classifier.predict(X_test)
```

## ***Step 10***: Predict The Output For Testing Dataset

```python
y_Predict=classifier.predict(X_test)
y_Predict
```

```python
print('Accuracy of the model is :',accuracy_score(y_Predict,y_test))
```

## ***Step 11***: Visualizing the Decision Tree Model

```python
# Visualizing the trained Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier,plot_tree
FEATURE_NAMES = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
TARGET_NAMES = ['SL','SW','PL','PW']
model=DecisionTreeClassifier(max_depth=4)
model.fit(X_train, y_train)
plt.figure(figsize=(12,6))
plot_tree(model,fontsize=10,feature_names=FEATURE_NAMES,class_names=TARGET_NAMES,rounded=True,filled=True)
plt.tight_layout()
plt.show()
```

## ***Step 12***: Calculating the Decision Tree Model accuracy

```python
# Model Accuracy, how often is the classifier correct?
print('Accuracy of the model is :',accuracy_score(y_Predict,y_test))
```
